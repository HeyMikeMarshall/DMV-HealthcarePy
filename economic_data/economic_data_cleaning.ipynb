```python
import pandas as pd
import numpy as py

income_data1 = pd.read_csv("income1.csv")

income_data1 = income_data1.loc[:,['Id2','Estimate; Per capita income in the past 12 months (in 2017 inflation-adjusted dollars)']]

income_data1.head()
counts = income_data1['Estimate; Per capita income in the past 12 months (in 2017 inflation-adjusted dollars)'].value_counts()
income_data1["Per Capita Income"] = pd.to_numeric(income_data1['Estimate; Per capita income in the past 12 months (in 2017 inflation-adjusted dollars)'].astype(str).str.replace(',',''), errors='coerce').fillna(0).astype(int)
income_data1 = income_data1.iloc[:,[0,2]]
income_data1["Per Capita Income"] = income_data1["Per Capita Income"].astype(py.int64)
income_data1 = income_data1.rename(columns={
    "Id2":"Zip Code"
})
```

```python
income_data2 = pd.read_csv("income2.csv")
income_data2 = income_data2.loc[:,['Id2','Total households - Less than $10,000',
                                   'Total households - \$10,000-$14,999',
                                   'Total households - \$15,000-$24,999',
                                   'Total households - \$25,000-$34,999',
                                   'Total households - \$35,000-$49,999',
                                   'Total households - \$50,000-$74,999',
                                   'Total households - \$75,000-$99,999',
                                   'Total households - \$100,000-$149,999',
                                   'Total households - \$150,000-$199,999',
                                   'Total households - \$200,000 or more']]
income_data2 = income_data2.rename(columns={"Id2":"Zip Code"})
income_data2.head()
```

```python
income_data3 = pd.read_csv("income3.csv")
income_data3 = income_data3.loc[:,["Id2",
                                   'Number; Estimate; Households',
                                   'Median income (dollars); Estimate; Households',
                                   'Number; Estimate; FAMILIES - Families',
                                   'Median income (dollars); Estimate; FAMILIES - Families',
                                   'Number; Estimate; NONFAMILY HOUSEHOLDS - Nonfamily households',
                                   'Median income (dollars); Estimate; NONFAMILY HOUSEHOLDS - Nonfamily households']]
income_data3 = income_data3.rename(columns={"Id2":"Zip Code",
                                   'Number; Estimate; Households':"Total Households",
                                   'Median income (dollars); Estimate; Households':"Median Income by Household",
                                   'Number; Estimate; FAMILIES - Families':"Total Families",
                                   'Median income (dollars); Estimate; FAMILIES - Families':"Median Income by Families",
                                   'Number; Estimate; NONFAMILY HOUSEHOLDS - Nonfamily households':"Total Nonfamily Households",
                                   'Median income (dollars); Estimate; NONFAMILY HOUSEHOLDS - Nonfamily households':"Median Income by Nonfamily Households"
    
})
```

```python
income_data3["Median Income by Household"] = pd.to_numeric(income_data3['Median Income by Household'].astype(str).str.replace(',',''), errors='coerce').fillna(0).astype(int)
income_data3["Median Income by Household"] = income_data3["Median Income by Household"].astype(py.int64)

income_data3["Median Income by Families"] = pd.to_numeric(income_data3['Median Income by Families'].astype(str).str.replace(',',''), errors='coerce').fillna(0).astype(int)
income_data3["Median Income by Families"] = income_data3["Median Income by Families"].astype(py.int64)

income_data3["Median Income by Nonfamily Households"] = pd.to_numeric(income_data3['Median Income by Nonfamily Households'].astype(str).str.replace(',',''), errors='coerce').fillna(0).astype(int)
income_data3["Median Income by Nonfamily Households"] = income_data3["Median Income by Nonfamily Households"].astype(py.int64)
```

```python
incomedata = pd.merge(income_data1, income_data2, how='outer', on='Zip Code')
incomedata = pd.merge(incomedata, income_data3,how="outer", on="Zip Code")
incomedata.head()
```

```python
census_data = pd.read_csv("census_data_uncleaned.csv")
census_data.head()
```

```python
censusdata = census_data.iloc[:,[1,3,5,9]]
censusdata = censusdata.rename(columns={
    "Id2":"Zip Code",
    "Total; Estimate; Population 16 years and over":"Total Population",
    "Labor Force Participation Rate; Estimate; Population 16 years and over":"Labor Force Participation Rate",
    "Unemployment rate; Estimate; Population 16 years and over":'Unemployment Rate'
})
censusdata["Total Population"] = pd.to_numeric(censusdata['Total Population'].astype(str).str.replace(',',''), errors='coerce').fillna(0).astype(int)
censusdata["Total Population"] = censusdata["Total Population"].astype(py.int64)

censusdata["Labor Force Participation Rate"] = pd.to_numeric(censusdata['Labor Force Participation Rate'].astype(str).str.replace(',',''), errors='coerce').fillna(0).astype(int)
censusdata["Labor Force Participation Rate"] = censusdata["Labor Force Participation Rate"].astype(py.int64)

censusdata["Unemployment Rate"] = pd.to_numeric(censusdata['Unemployment Rate'].astype(str).str.replace(',',''), errors='coerce').fillna(0).astype(int)
censusdata["Unemployment Rate"] = censusdata["Unemployment Rate"].astype(py.int64)

censusdata.dtypes
```

```python
econ_data = pd.merge(censusdata,incomedata,how="left",on="Zip Code")
econ_data.head()
econ_data.dtypes
```

```python
econ_data.to_csv(r'..\all_econ_data.csv', index = None, header=True)
```
