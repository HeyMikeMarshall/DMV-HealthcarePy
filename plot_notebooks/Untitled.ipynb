```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("primary_data.csv")
data.head()
```

```python
data["Per Capita Income"].max()

plt.hist(data["Per Capita Income"], bins=10)
plt.ylabel('No of times')
plt.show()

data.head()
```

```python
# Sample bins.
income_bins = py.array([0, 10000,20000,30000, 40000,50000 ,60000, 70000,80000, 90000,100000,120000])
bin_names = ["0, 10000","10000,20000","20000,30000","30000, 40000","40000,50000","50000 ,60000","60000, 70000","70000,80000","80000, 90000"," 90000,100000","100000,120000"]

#Create new DataFrame and bin
bindata = data
bindata["Inc Group"] = pd.cut(bindata["Per Capita Income"], income_bins, labels=bin_names)

#Group by budgets and find value means
new = bindata.groupby(['Inc Group']).sum()#.reset_index()

new.plot.bar(y="Total Population", figsize=(7,4))
data["Per Capita Income"].max()
new.head()
```
