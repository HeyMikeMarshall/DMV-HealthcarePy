```python
import pandas as pd
from matplotlib.cm import viridis
from matplotlib.colors import to_hex
import matplotlib.pyplot as plt
import numpy as np
import ast
import json
```

```python
primary_data_to_load = r'/Users/VShekhar/gitrepositories/DMV-HealthcarePy/primary_data.csv'
primary_data_df = pd.read_csv(primary_data_to_load)
new_df = primary_data_df
new_df
```

```python

frame_test = new_df.head(100)
frame = frame_test.loc[:,["Unemployment Rate","hospital facilities", "doctors"]].set_index('Unemployment Rate')
frame2 = frame.sort_values(by=['Unemployment Rate'])
frame2.plot(kind='bar', stacked=True, figsize=(15, 12))
```

```python
frame9 = new_df.loc[:,["Total Population","hospital facilities", "doctors"]].set_index('Total Population')
frame10 = frame9.sort_values(by=['Total Population'])
frame10.plot(kind='bar', stacked=True, figsize=(15, 12))
```

```python
frame3 = new_df.loc[:,["Unemployment Rate","hospital facilities", "doctors"]]
plt.scatter(frame3["Unemployment Rate"], frame3["hospital facilities"], color = "red", marker = "o", alpha = 0.5, linewidth = 2.1)
plt.scatter(frame3["Unemployment Rate"], frame3["doctors"], color = "blue", marker = "o", alpha = 0.5, linewidth = 2.1)

plt.title("Doctors and Hospital Facilities vis-a-vis Unemployment Rate")
plt.xlabel("Unemployment Rate")

# Create a legend.
plt.legend(loc= "upper right")

# Show the Plot

plt.show()
```

```python
frame3 = new_df.loc[:,["Total Population","hospital facilities", "doctors"]]
plt.scatter(frame3["Total Population"], frame3["hospital facilities"], color = "green", marker = "o", alpha = 0.5, linewidth = 2.1)
plt.scatter(frame3["Total Population"], frame3["doctors"], color = "orange", marker = "o", alpha = 0.5, linewidth = 2.1)
plt.title("Doctors and Hospital Facilities vis-a-vis Total Population")
plt.xlabel("Total Population")

# Create a legend.
plt.legend(loc= "upper right")

# Show the Plot
plt.show()

```

```python
frame3 = new_df.loc[:,["Per Capita Income","hospital facilities", "doctors"]]
plt.scatter(frame3["Per Capita Income"], frame3["hospital facilities"], color = "green", marker = "o", alpha = 0.5, linewidth = 2.1)
plt.scatter(frame3["Per Capita Income"], frame3["doctors"], color = "orange", marker = "o", alpha = 0.5, linewidth = 2.1)
plt.title("Doctors and Hospital Facilities vis-a-vis Per Capita Income")
plt.xlabel("Per Capita Income")

# Create a legend.
plt.legend(loc= "upper right")

# Show the Plot
plt.show()
```
